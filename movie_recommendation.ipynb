```python
  #Importing Library
import numpy as np
import pandas as pd
import random
```

## feature selection

```python
movie = movie[['MOVIES','YEAR','GENRE','RATING','ONE-LINE','STARS','VOTES','RunTime','Gross']]
```

#Filling the null value of **RUNTIME**

```python
movie['RunTime']=movie['RunTime'].replace(np.nan, random.choice(['120','80',]))
```

#Assigning random RATES too

```python
movie['RATING']=movie['RATING'].replace(np.nan, random.choice(['5.1','7.2','5.7','6.5']))
```

# removing **\n** from start of all the GENRES

```python
movie['GENRE'] = movie['GENRE'].str.strip()
movie
```

#Checking how many **null** values are present in each **feature**

```python
missing_values_count = movie.isnull().sum()
print(missing_values_count)
```

#**GROSS** has too many Null Values so better **DROP** it

```python
movie = movie.drop(columns=['Gross'])
```

#Modification of RunTime feature

```python
movie.loc[movie['GENRE'] == 'Comedy', 'VOTES'] = movie.loc[movie['GENRE'] == 'Comedy', 'VOTES'].fillna(1200)
movie.loc[movie['GENRE'] == 'Thriller', 'VOTES'] = movie.loc[movie['GENRE'] == 'Thriller', 'VOTES'].fillna(1800)
movie.loc[movie['GENRE'] == 'Action', 'VOTES'] = movie.loc[movie['GENRE'] == 'Action', 'VOTES'].fillna(1600)
#didNotReallyWorked...
```

```python
def Convert(string):
    if isinstance(string, str):
        return string.split(",")
    return []
```

```python
#def Convert(string):
#    li = (string.split(","))
#    return li
movie['GENRE']=movie['GENRE'].apply(Convert)
movie['GENRE']
```

```python
movie['GENRE']=movie['GENRE'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie
```

Making list of all other features too so that tags can be generated

```python
movie['ONE-LINE']=movie['ONE-LINE'].apply(Convert)
movie['STARS']=movie['STARS'].apply(Convert)
movie['YEAR']=movie['YEAR'].apply(Convert)
movie['RunTime']=movie['RunTime'].apply(Convert)
movie['RATING']=movie['RATING'].apply(Convert)
```

```python
movie
```

left the **VOTES** features as too much nulls

```python
movie['Movie_name'] = movie['MOVIES']
movie['tags']= movie['GENRE']+movie['RATING']+movie['ONE-LINE']+movie['STARS']+movie['RunTime']
```

```python
movie
```

```python
new_df=movie[['Movie_name','tags']]
new_df['tags'] =new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
def Convert(string):
    li = string.lower()
    return li
```

```python
new_df['Movie_name']=new_df['Movie_name'].str.lower()
```

```python
new_df
```

```python

#####Imporint feature extractor################

import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(new_df['tags']).toarray()


print(vector)
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
  y =[]

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)


new_df['tags']=new_df['tags'].apply(stem)
print(new_df)
similarity = cosine_similarity(vector)
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:6]
similarity[0]
```

```python
######OUTPUT####
def recommend(movie):
    movie_index = new_df[new_df['Movie_name'] == movie].index[0]
    distances = similarity[movie_index]
    movie = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in movie:
        print(new_df.iloc[i[0]].Movie_name)
recommend('the walking dead')
```
